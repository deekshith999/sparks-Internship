# Prediction using Decision Tree Algorithm

### Decision Tree Classifier

Create the Decision Tree classifier and visualize it graphically.

### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

### Importing Dataset

```python
df=pd.read_csv("Iris.csv")
print('Data Downloaded')
df.head()
```

### Preparing Dataset

```python
X=df.iloc[:,1:5].values
Y=df.iloc[:,5].values
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

### Using Decision Tree Classifier

```python
dtmodel=DecisionTreeClassifier()
dtmodel.fit(X_train,Y_train)
```

```python
Y_pred=dtmodel.predict(X_test)
```

### Evaluating the Classifier

```python
from sklearn import metrics
print('Precision: {}'.format(metrics.precision_score(Y_test,Y_pred,average='macro')))
print('Recall: {}'.format(metrics.recall_score(Y_test,Y_pred,average='macro')))
print('Jaccard Score: {}'.format(metrics.jaccard_score(Y_test,Y_pred,average='macro')))
print('F1 Score: {}'.format(metrics.f1_score(Y_test,Y_pred,average='macro')))
```

### Visualizing the Tree

```python
plt.figure(figsize=(15,10))
tree.plot_tree(dtmodel,feature_names=df.iloc[:,1:5].columns.values,class_names=df.Species.unique(),filled=True)
plt.title('Iris Dataset Decision Tree')
plt.show()
```
